# 1. Importing necessary libraries

```python
import string
import re
import fileinput
```

# 2. Specifying Input and Output file paths

# 3. Tokenisation and Normalisation(Case Folding)

```python
# print(string.punctuation)
def tokenizing_func(input_file, output_file, lower_cased):
    # Read the content of the file to be tokenized and normalised 
    with open(input_file, 'r') as file1:
        content1 = file1.read()

    # Remove Punctuations
    for char in content1:
        if char in string.punctuation:
            content1 = content1.replace(char, '')

    with open(output_file, 'w') as file2:
        file2.write(content1)    

    with open(output_file, 'r') as file:
        to_lower= file.read()


    # NORMALISATION(Case Folding)
    with open(lower_cased, 'w') as file:
        file.write(to_lower.lower())
```

# 4. Removing Stopwords

```python
def remove_stop_words(file1_path, file2_path, no_stopwords):
    # Read the content of the first file
    with open(file1_path, 'r') as file1:
        content1 = file1.read()
    # Read the content of the second file
    with open(file2_path, 'r') as file2:
        content2 = file2.read()
    
    # Split the content into lists of words
    words1 = content1.split()
    words2 = content2.split()

    # Find words in words1 that are not in words2 while maintaining the order
    unique_words = [word for word in words1 if word not in words2]

    # Write the unique words(not in stopwords list) to the output file
    with open(no_stopwords, 'w') as file:
        file.write("\n".join(unique_words))
        
```

# 5. Stemming

```python
from stemming.porter2 import stem
def stem_words(without_stpwrds, file_path_stem, file_path_stem_unique ):
    # Read the content of the file
    with open(without_stpwrds, 'r') as file:
        content = file.read()

    words = content.split()

    # Stem each word
    stemmed_words = [stem(word) for word in words]

    # Join the stemmed words back into a string
    stemmed_content = '\n'.join(stemmed_words)

    # Write the stemmed content back to the file
    with open(file_path_stem, 'w') as file:
        file.write(stemmed_content)

     # Remove duplicates by converting the list to a set, then convert it back to a list
    unique_stemmed_words = list(dict.fromkeys(stemmed_words))

     # Join the stemmed words back into a string
    unique_stemmed_content = '\n'.join(unique_stemmed_words)

    # Write the unique stemmed content back to the file
    with open(file_path_stem_unique, 'w') as file:
        file.write(unique_stemmed_content)
```

# PREPROCESSING

1. Quran

```python
stopwords_file_path = 'stopWords.txt'
input_file_path = 'quran.txt'  
removed_stopwords="Quran/Quran_without_stopwords.txt"
output_file_path = "Quran/tokenized_Quran.txt"  
lower_cased_filepath="Quran/lower_cased_Quran.txt"
# For stemming
file_path_stem = 'Quran/stem_Quran_without_stopwords.txt'
# For Unique stem words
file_path_stem_unique = 'Quran/unique_stemmed_Quran.txt'
# Remove punctuation from the input file and save to the output file
tokenizing_func(input_file_path, output_file_path,lower_cased_filepath)
# Compare the content of the specified files and remove stopwords
remove_stop_words(lower_cased_filepath, stopwords_file_path,removed_stopwords)

#remove numbers from the bible
for line in fileinput.input(removed_stopwords, inplace=True):
    print(re.sub("\d+", "", line))

# Stem the words in the specified file
stem_words(removed_stopwords,file_path_stem, file_path_stem_unique)
print("Quran Successfully preprocessed!")

```

# ZIPF'S LAW

1. Quran

```python
import collections
import matplotlib.pyplot as plt
import numpy as np

def plot_zipfs_law(file_path):
    with open(file_path, 'r') as file:
        words = file.read().split()

    # Count the frequency of each word
    word_counts = collections.Counter(words)

    # Sort the words by their frequency
    sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

    # Print each word and its frequency
    for word, count in sorted_word_counts:
        print(f"{word}: {count}")

    # Extract the frequencies and ranks
    frequencies = [count for word, count in sorted_word_counts]
    ranks = range(1, len(frequencies) + 1)

    # Plot the frequencies in a log-log graph
    plt.plot(ranks, frequencies)
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.title('Zipf\'s Law')
    plt.show()
# Plot Zipf's law for each collection
plot_zipfs_law('Quran/stem_Quran_without_stopwords.txt')
```

# HEAP'S LAW

1. Quran

```python
import matplotlib.pyplot as plt
import numpy as np


def plot_heaps_law(file_path):
    with open(file_path, 'r') as file:
        words = file.read().split()

    # Initialize the vocabulary and the lists to store the number of terms and the vocabulary size
    vocabulary = set()
    num_terms = []
    vocab_sizes = []

    # Go through the words one by one
    for i, word in enumerate(words):
        vocabulary.add(word)
        num_terms.append(i + 1)
        vocab_sizes.append(len(vocabulary))
    # print("Our Vocabularies are: ",vocabulary)
    print("The vocabulary size is: ",vocab_sizes[-1])
    print("The number of terms is: ",num_terms[-1])
    

    # Plot the growth of vocabulary
    plt.plot(num_terms, vocab_sizes)
    plt.xlabel('Number of Terms')
    plt.ylabel('Vocabulary Size')
    plt.show()

    # Fit Heap's law to the data and print the best fitting k and b
    log_num_terms = np.log(num_terms)
    log_vocab_sizes = np.log(vocab_sizes)
    b, log_k = np.polyfit(log_num_terms, log_vocab_sizes, 1)
    k = np.exp(log_k)
    print(f"Best fitting k: {k}, b: {b}")

# Plot Heap's law for each collection
plot_heaps_law('Quran/stem_Quran_without_stopwords.txt')
```

# BENFORD'S LAW(Frequencies)

```python
import collections
import matplotlib.pyplot as plt
import numpy as np

def plot_benfords_law(file_path):
    with open(file_path, 'r') as file:
        words = file.read().split()

    # Count the frequency of each word
    word_counts = collections.Counter(words)
    # print("The word counts are: ", word_counts)

    # Get the first digit of each frequency
    first_digits = [int(str(count)[0]) for count in word_counts.values() if count >= 10]
    # print("The first digits are: ", first_digits)

    # Count the frequency of each first digit
    digit_counts = collections.Counter(first_digits)
    print("The digit counts are: ", digit_counts)

    # Plot the distribution of the first digits
    digits, frequencies = zip(*sorted(digit_counts.items()))
    
    plt.bar(digits, frequencies)
    plt.xlabel('First Digit')
    plt.ylabel('Frequency')
    plt.show()

# Plot Benford's law for each collection
plot_benfords_law('Quran/stem_Quran_without_stopwords.txt')
```

# BENFORD'S LAW (probabilities)

```python
import collections
import matplotlib.pyplot as plt

def plot_benfords_law(file_path):
    with open(file_path, 'r') as file:
        words = file.read().split()

    word_counts = collections.Counter(words)

    first_digits = [int(str(count)[0]) for count in word_counts.values() if count >= 10]

    digit_counts = collections.Counter(first_digits)

    total_count = sum(digit_counts.values())
    digit_probabilities = {digit: count / total_count for digit, count in digit_counts.items()}
    print("The digit probabilities are: ",  digit_probabilities)

    digits, probabilities = zip(*sorted(digit_probabilities.items()))
    print("The probabilities are: ",  probabilities)
    print("The digit are: ",  digits)


    plt.bar(digits, probabilities)
    plt.xlabel('First Digit')
    plt.ylabel('Probability')
    plt.title("Benford's Law")
    plt.show()

# Example usage
plot_benfords_law('Quran/stem_Quran_without_stopwords.txt')
```
